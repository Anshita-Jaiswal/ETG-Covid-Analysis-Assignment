## Q1. Total number of people vaccinated in Argentina

```python
temp = df[df['country']=='Argentina']
temp
```

```python
temp = temp['people_vaccinated']
temp
```

```python
temp.sort_values(ascending = False)
```

```python
temp.max()
# total number of people vaccinated in Argentina
```

## Q3. Maximum number of people vaccinated in a country

```python
df.head()
```

```python
df.groupby('country').max()['people_vaccinated']
```

```python
top = df.groupby('country').max()['people_vaccinated'].sort_values(ascending = False)
top
```

```python
top[:1]
# maximum number of people vaccinated in a country 
```

```python
top.max()
# maximum number of people vaccinated in a country
```

## Q4. Top 5 Countries in terms of number of people vaccinated

```python
df.groupby('country').max()['people_vaccinated'].sort_values(ascending = False).head(5)
# top 5 countries in terms of number of people vaccinated
```
